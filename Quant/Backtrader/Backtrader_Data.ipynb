## 1. From Tushare

```python
import tushare as ts

ts.set_token('token')
pro = ts.pro_api()

start_date = '20200101' 
end_date = '20240131' 

df = pro.daily(ts_code='000001.SZ', start_date=start_date, end_date=end_date)

df
```

## 2. From Baostock

```python
import baostock as bs


bs.login()

df = bs.query_history_k_data_plus("sh.600000",
    "date,code,open,high,low,close,volume",
    start_date="2020-01-01", end_date="2024-01-31",
    frequency="d", adjustflag="3").get_data()


bs.logout()

df
```
